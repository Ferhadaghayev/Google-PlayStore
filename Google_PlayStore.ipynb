```python
import pandas as pd
df=pd.read_csv('/content/googleplaystore.csv')
```

```python
#int tipinə çevir
#df['Rating']=pd.to_numeric(df['Rating'],errors='coerce')
#df['Price']=pd.to_numeric(df['Price'],errors='coerce')
df['Reviews']=pd.to_numeric(df['Reviews'],errors='coerce')
```

```python
df
```

```python
df.describe()
```

```python
#Size sütununda M->0 əvəz edir
df['Size']=df['Size'].str.replace('M','0000')
```

```python
#Unique dəyərlər - Unique values
df.nunique()
```

```python
#NULL dəyərlər -- NULL values
df.isnull().sum()
```

```python
#Sütundakı nan dəyərləri doldururuq -- Filling in the nan values ​​in the column
df['Type'] = df['Type'].fillna(0)
```

```python
df['Installs'].value_counts()
```

```python
df['Rating'].unique()
```

```python
# df['Price']=df['Price'].astype(int)
import numpy as np
df['Price']=df['Price'].replace('Everyone',np.nan)
```

```python
# df['Price']=df['Price'].fillna(df['Price'].mean())
df['Price'].nunique()
```

```python
# $ işarəsini sil -- deleting $
df['Price']=df['Price'].str.replace('$','')
```

```python
df['Rating'] = df['Rating'].astype(float)
df['Rating'] = df['Rating'].replace(19. , df['Rating'].mean())
```

```python
df['Rating'].value_counts()
```

```python
df['Rating'].unique()
```

```python
#Histogramla ümumi rating paylanması -- Overall rating distribution with histogram
import matplotlib.pyplot as plt

df['Rating'].hist()
```

```python

df['Android Ver']=pd.to_numeric(df['Android Ver'],errors='coerce')
```

```python
df['Android Ver']=df['Android Ver'].fillna(df['Android Ver'].mean())
```

```python
df['Rating']=df['Rating'].fillna(df['Rating'].mean())
```

```python
df['Current Ver']=df['Current Ver'].fillna(df['Current Ver'].mode()[0])
```

```python
df['Price']=df['Price'].fillna(df['Price'].mode()[0])
```

```python
df['Content Rating']=df['Content Rating'].fillna(df['Content Rating'].mode()[0])
```

```python
df['Reviews']=df['Reviews'].fillna(df['Reviews'].mode()[0])
```

```python
df=df.drop(['Android Ver'],axis=1)
```

```python
df.isnull().sum()
```

```python
df['Last Updated']=pd.to_datetime(df['Last Updated'],errors='coerce',format='mixed')
```

```python
corr=df[['Rating','Price']].corr()
corr
```

```python
#Qiymət vs Reytinq Qrafiki - Price vs Rating Chart
import seaborn as sns
sns.heatmap(corr,annot=True)
```

```python
plt.figure(figsize=(20,15))
plt.hist(df['Genres'],bins=(8))
plt.title("G")
plt.xlabel("Reviews")
plt.ylabel("App")
plt.show()
```

```python
#Application-ların Rating paylanması
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df['Rating'].dropna(), bins=20)
plt.xlabel("Rating")
plt.ylabel("Num of Apps-App sayı")
plt.title("Distribution of App Ratings")
plt.show()
```

```python
#Kateqoriya üzrə reytinq paylanması
category_rating = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
category_rating.plot(kind='bar')
plt.ylabel("Average Rating")
plt.title("Average Rating by Category")
plt.show()
```

```python
#Free & Paid(Ödənişli) Apps
type_counts = df['Type'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%')
plt.title("Free & Paid(Ödənişli) Apps")
plt.show()
```

```python
#Category üzrə App sayı-Count of Apps by Category
category_counts = df['Category'].value_counts()

plt.figure(figsize=(8,7))
plt.barh(category_counts.index, category_counts.values)
plt.xlabel("Number of Apps")
plt.title("Number of Apps by Category")
plt.show()
```

```python
#Ödənişli olan App-lərdə qiymət paylanması
paid_apps = df[df['Type'] == 'Paid']

plt.figure(figsize=(8,5))
plt.hist(paid_apps['Price'], bins=30)
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Price Distribution of Paid Apps")
plt.show()
```
